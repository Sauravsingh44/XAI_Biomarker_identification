```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
from sklearn.model_selection import train_test_split
```

```python
file_path = r"C:\Users\Saurav\Documents\Colorectal_cancer\filtered_ml_dataset.csv"
df = pd.read_csv(file_path)
df
```

```python
df1 = df.drop(columns=["Sample"])
df1
```

```python
X = df1.drop(columns="Condition")
y = df1["Condition"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3 ,random_state=42)
```

```python
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_train)
accuracy = accuracy_score(y_pred,y_train)
print(f"Training accuracy:{accuracy*100:.3f}%")
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Testing accuracy: {test_accuracy * 100:.3f}%")
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Your data
actual = y_test
predicted = model.predict(X_test)

# Compute confusion matrix
cm = confusion_matrix(actual, predicted)

# Plot the confusion matrix
sns.heatmap(cm, annot=True, fmt='g',
            xticklabels=['Not Infected','Infected'],
            yticklabels=['Not Infected','Infected'])
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.title("Confusion Matrix")
plt.gca().xaxis.set_label_position('top')
plt.gca().xaxis.tick_top()
plt.gca().figure.subplots_adjust(bottom=0.2)
plt.show()

# Extract values from the confusion matrix
TP = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TN = cm[1][1]

# Calculate metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
recall = TP / (TP + FN) if (TP + FN) != 0 else 0  # Sensitivity
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

# Print the results
print(f"Accuracy: {accuracy*100:.2f}")
print(f"Precision: {precision*100:.2f}")
print(f"Sensitivity (Recall): {recall*100:.2f}")
print(f"F1-Score: {f1_score*100:.2f}")
```

```python
from sklearn.metrics import roc_curve, auc, accuracy_score
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class (index 1)
y_prob = model.predict_proba(X_test)[:, 1]

# Predict class labels
y_pred = model.predict(X_test)

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'ROC Curve (SVM)\nAccuracy: {accuracy * 100:.2f}%')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

# SVM using grid search

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.svm import SVC
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
			'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
			'kernel': ['rbf'],
             'probability':[True]} 

grid = GridSearchCV(svm.SVC(), param_grid, refit = True, verbose = 3) 
 
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_) 
 
print(grid.best_estimator_)
```

```python
from sklearn.metrics import accuracy_score
y_pred = grid.predict(X_train)
accuracy = accuracy_score(y_pred,y_train)
print(f"Training accuracy:{accuracy*100:.3f}%")
```

```python
from sklearn.metrics import accuracy_score

y_pred = grid.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Testing accuracy: {test_accuracy * 100:.3f}%")
```
