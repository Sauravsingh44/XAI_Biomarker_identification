```python
import pandas as pd

# Load the DEGs dataset
# Replace 'degs_file.csv' with your actual file name
degs_df = pd.read_csv('combined_DEGs.csv')

# Load the ML_dataset
# Replace 'ml_file.csv' with your actual file name
ml_df = pd.read_csv('ML_dataset.csv')
```

```python
import pandas as pd

# Assume you've already loaded your dataframes as degs_df and ml_df
# Example (if you haven't loaded them yet):
# degs_df = pd.read_csv('degs.csv')
# ml_df = pd.read_csv('ml.csv')

# Extract symbols from degs_df
degs_symbols = set(degs_df['SYMBOL'])

# Extract symbols from ml_df.
# We'll get all column names and then remove the 'Sample' column.
ml_symbols = set(ml_df.columns) - {'Sample'}

# Now you have two sets: one containing all symbols from degs_df,
# and another containing all gene symbols from ml_df.
```

```python
# Find the intersection of the two sets
common_symbols = degs_symbols.intersection(ml_symbols)

# Print the number of common symbols
print(f"Number of common symbols found: {len(common_symbols)}")

# Optionally, print the list of common symbols
print("List of common symbols (first 10):")
print(list(common_symbols)[:10])
```

```python
# Filter ml_df using the common symbols
columns_to_keep = ['Sample'] + list(common_symbols) + ['Condition']
ml_common_df = ml_df[columns_to_keep]

print("Filtered ml_df with only common symbols:")
ml_common_df
```

```python
ml_common_df['Condition'] = ml_common_df['Condition'].replace({ 'normal' : 0, 'colorectal cancer' : 1})
ml_common_df
```

```python
# Assuming ml_common_df is your DataFrame
ml_common_df.to_csv('filtered_ml_dataset.csv', index=False)
```
