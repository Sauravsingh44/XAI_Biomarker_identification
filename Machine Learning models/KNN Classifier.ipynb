```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import os
```

```python
file_path = r"C:\Users\Saurav\Documents\Colorectal_cancer\filtered_ml_dataset.csv"
df = pd.read_csv(file_path)
df
```

```python
df1 = df.drop(columns=["Sample"])
df1
```

```python
from sklearn.model_selection import train_test_split
X = df1.drop(columns="Condition")
y = df1["Condition"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3 ,random_state=42)
```

# K-NN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
k_values=range(1,21)
error_rate=[]

for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    error=1-accuracy_score(y_pred,y_test)
    error_rate.append(error)

plt.figure(figsize=(8, 6))
plt.plot(k_values, error_rate, marker='o')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
plt.title('Elbow Method for Finding Optimal K')
plt.axvline(x=3, linestyle='--', color='red', label="Optimal K (Elbow Point)")
plt.legend()
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
accuracy=accuracy_score(y_pred,y_test)
print(f"Testing Accuracy:{accuracy*100:.2f}%")
```

```python
y_pred=knn.predict(X_train)
accuracy=accuracy_score(y_pred,y_train)
print(f"Training Accuracy:{accuracy*100:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

actual=y_test
predicted=knn.predict(X_test)
cm=confusion_matrix(actual,predicted)

sns.heatmap(cm,annot=True,fmt='g',
           xticklabels=['Not Infected','Infected'],
           yticklabels=['Not Infected','Infected'])
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.title("Confusion Matrix")
plt.gca().xaxis.set_label_position('top') 
plt.gca().xaxis.tick_top()
plt.gca().figure.subplots_adjust(bottom=0.2)
plt.show()


TP = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TN = cm[1][1]

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
recall = TP / (TP + FN) if (TP + FN) != 0 else 0  
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Sensitivity (Recall): {recall*100:.2f}%")
print(f"F1-Score: {f1_score*100:.2f}%")
```

```python
print(classification_report(predicted,actual))
```

```python
from sklearn.metrics import roc_curve, auc, accuracy_score
import matplotlib.pyplot as plt

# Predict probabilities (only for the positive class)
y_prob = knn.predict_proba(X_test)[:, 1]

# Predict labels (for accuracy)
y_pred = knn.predict(X_test)

# Calculate ROC metrics
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve\nAccuracy: {:.2f}%'.format(accuracy * 100))
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
